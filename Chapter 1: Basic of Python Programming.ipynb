1. Write a program to calculate bodymass index BMI of a person. (bmi = kg/m^2)

```python
w = float(input("Input your weight in kg: "))
h = float(input("Input your height in cm: "))

hm = h / 100

bmi = w / (hm * hm)

print(f"Your BMI is {bmi:.2f}")
```
